```python
# 2022-01-01 ~ 2024-12-31 모든 회사의 모든 공시목록
df1 = dart.list(start='2022-01-01', end='2022-03-31')
df2 = dart.list(start='2022-04-01', end='2022-06-30')
df3 = dart.list(start='2022-07-01', end='2022-09-30')
df4 = dart.list(start='2022-10-01', end='2022-12-31')
df5 = dart.list(start='2023-01-01', end='2023-03-31')
df6 = dart.list(start='2023-04-01', end='2023-06-30')
df7 = dart.list(start='2023-07-01', end='2023-09-30')
df8 = dart.list(start='2023-10-01', end='2023-12-31')
df9 = dart.list(start='2024-01-01', end='2024-03-31')
df10 = dart.list(start='2024-04-01', end='2024-06-30')
df11 = dart.list(start='2024-07-01', end='2024-09-30')
df12 = dart.list(start='2024-10-01', end='2024-12-31')
```

```python
df1_ri = df1.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
df2_ri = df2.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
df3_ri = df3.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
df4_ri = df4.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
df5_ri = df5.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
df6_ri = df6.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
df7_ri = df7.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
df8_ri = df8.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
df9_ri = df9.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
df10_ri = df10.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
df11_ri = df11.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
df12_ri = df12.query("report_nm in ['[기재정정]주요사항보고서(유상증자결정)', '주요사항보고서(유상증자결정)']")
```

```python
corp_name = [] # corp_name : 회사명
date = [] # date : 날짜
common_new = [] # common_new : 신주의 수(보통주식)
other_new = [] # other_new : 신주의 수(기타주식) 
for_facility = [] # for_facility : 자금조달목적=시설자금, 필요한 금액
for_acquisition = [] # for_acquisition : 자금조달목적=영업양수자금, 필요한 금액
for_manage = [] # 운영자금
for_repay = [] # 채무상환자금
for_corp = [] # 타법인 증권 취득자금
for_other = [] # 기타자금
way = [] # way : 증자방식

dfs = [df1_ri, df2_ri, df3_ri, df4_ri, df5_ri, df6_ri, df7_ri, df8_ri, df9_ri, df10_ri, df11_ri, df12_ri]


for df in dfs:
    for _, r in df.iterrows():
        detail = dart.event(r['corp_code'], '유상증자')
        dum = detail[detail['rcept_no'] == r['rcept_no']]

        if not dum.empty:
            corp_name.append(dum['corp_name'].iloc[0])
            date.append(dum['rcept_no'].iloc[0][:9])
            common_new.append(dum['nstk_ostk_cnt'].iloc[0])
            other_new.append(dum['nstk_estk_cnt'].iloc[0])
            for_facility.append(dum['fdpp_fclt'].iloc[0])
            for_acquisition.append(dum['fdpp_bsninh'].iloc[0])
            for_manage.append(dum['fdpp_op'].iloc[0])
            for_repay.append(dum['fdpp_dtrp'].iloc[0])
            for_corp.append(dum['fdpp_ocsa'].iloc[0])
            for_other.append(dum['fdpp_etc'].iloc[0])
            way.append(dum['ic_mthn'].iloc[0])
        else:
            corp_name.append(None)
            date.append(None)
            common_new.append(None)
            other_new.append(None)
            for_facility.append(None)
            for_acquisition.append(None)
            for_manage.append(None)
            for_repay.append(None)
            for_corp.append(None)
            for_other.append(None)
            way.append(None)

```

```python
import pandas as pd
df_final = pd.DataFrame({'corp_name': corp_name, 'date': date, 'common_new':common_new, 'other_new': other_new, 'for_facility': for_facility,
                          'for_acquisition': for_acquisition, 'for_manage': for_manage, 'for_repay': for_repay,
                          'for_corp': for_corp, 'for_other': for_other, 'way': way})
df_final
```

```python
df_final.to_excel("유상증자정리.xlsx", index=False)
```

```python
df.to_excel("유상증자정리(최종).xlsx", index=False)
```
